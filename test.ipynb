```python
# Test Step 4: Create time-based train/val/test split
from src.qaware.sequences import create_time_based_split

print("Creating train/val/test split...")
train_df, val_df, test_df = create_time_based_split(interactions_remapped, cfg)

print(f"Train interactions: {len(train_df)}")
print(f"Val interactions: {len(val_df)}")
print(f"Test interactions: {len(test_df)}")
print(f"\nTotal after split: {len(train_df) + len(val_df) + len(test_df)}")
```

```python
# Inspect a specific user's split
example_user_id = train_df["user_id"].iloc[0]
print(f"Example user: {example_user_id}")

user_train = train_df[train_df["user_id"] == example_user_id]
user_val = val_df[val_df["user_id"] == example_user_id]
user_test = test_df[test_df["user_id"] == example_user_id]

print(f"  Train: {len(user_train)} interactions")
print(f"  Val: {len(user_val)} interactions")
print(f"  Test: {len(user_test)} interactions")

# Show the sequence
print(f"\n  Train items (first 10): {user_train['product_id'].head(10).tolist()}")
if len(user_val) > 0:
    print(f"  Val item: {user_val['product_id'].iloc[0]}")
if len(test_df) > 0:
    print(f"  Test item: {user_test['product_id'].iloc[0]}")
```

```python
# Verify no data leakage (check that train/val/test don't overlap)
train_users = set(train_df["user_id"].unique())
val_users = set(val_df["user_id"].unique())
test_users = set(test_df["user_id"].unique())

print("User overlap check:")
print(f"  Users in train: {len(train_users)}")
print(f"  Users in val: {len(val_users)}")
print(f"  Users in test: {len(test_users)}")
print(f"  All users appear in all splits: {train_users == val_users == test_users}")

# Check that each user has exactly 1 val and 1 test interaction
val_counts = val_df.groupby("user_id").size()
test_counts = test_df.groupby("user_id").size()
print(f"\n  Users with exactly 1 val interaction: {(val_counts == 1).sum()}/{len(val_counts)}")
print(f"  Users with exactly 1 test interaction: {(test_counts == 1).sum()}/{len(test_counts)}")
```
